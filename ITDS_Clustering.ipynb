```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

plt.style.use('seaborn')
x,y = make_blobs(n_samples = 200, n_features = 2, centers = 4, cluster_std=0.5, shuffle = True, random_state=0)

plt.scatter(x[:,0],x[:,1],c='k', marker='+',s=50)
plt.show()
```

```python
y
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=4, init='random', n_init=10, max_iter=300, tol=1e-04, random_state=0)
y_km = km.fit_predict(x)

plt.scatter(x[y_km==0,0],x[y_km==0,1],c='r', marker='s',s=40, label='Cluster 1')
plt.scatter(x[y_km==1,0],x[y_km==1,1],c='g', marker='o',s=40, label='Cluster 2')
plt.scatter(x[y_km==2,0],x[y_km==2,1],c='b', marker='v',s=40, label='Cluster 3')
plt.scatter(x[y_km==3,0],x[y_km==3,1],c='y', marker='p',s=40, label='Cluster 4')

plt.scatter(km.cluster_centers_[:,0],km.clusters_centers_[:,1],c='k', marker='X',s=40, label='Centers')

plt.legend(scatterpoints=1)

plt.show()
```

```python
inertia = []

for i in range(1,15):
  km = KMeans(n_clusters=i, init='random', n_init=10, max_iter=300, tol=1e-04, random_state=0)
  # y_km = km.fit_predict(x)
  km.fit(x)
  inertia.append(km.inertia_)

plt.plot(range(1,15),inertia, marker='o')
plt.show
```

```python
from sklearn.datasets import make_moons

plt.style.use('seaborn')
x,y = make_moons(n_samples = 200, noise=0.05, shuffle = True, random_state=0)

plt.scatter(x[:,0],x[:,1],c='k', marker='+',s=50)
plt.show()
```

```python
from sklearn.datasets import make_circles

plt.style.use('seaborn')
x,y = make_circles(n_samples = 200, noise=0, shuffle = True, random_state=0)

plt.scatter(x[:,0],x[:,1],c='k', marker='+',s=50)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
newX = scaler.fit_transform(x)
dbscan = DBSCAN(eps=0.35,min_samples=3)

y_db = dbscan.fit_predict(newX)
plt.scatter(newX[:,0],newX[:,1],c=y_db, cmap='plasma')
plt.show()
```
